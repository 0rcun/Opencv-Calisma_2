```python
import cv2
import numpy
import time
import matplotlib.pylab as plt
%matplotlib inline
```

```python
vcap = cv2.VideoCapture(0)
```

```python
resimler=[]

for i in range(5):
    ret,image = vcap.read()
    resimler.append(image)
    isim="sn"+str(i)+".jpg"
    cv2.imwrite(str(isim),image)
    time.sleep(1)
for resimm in resimler:
    plt.imshow(resimm)
    plt.show()
```
